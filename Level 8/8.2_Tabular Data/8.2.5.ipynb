```python
# c. Display the total volume
np.sum(tsla.Volume)
```

```python
# d. Display the average daily volume
np.mean(tsla.Volume)
```

```python
# e. Add a column containing the daily ‘return'

tsla['Daily_return'] = (tsla['Volume']-tsla['Volume'].shift(1))/tsla['Volume'].shift(1)


tsla.head()
```

```python
tsla.tail()
```

```python
# f. Calculate the average daily return
np.mean(tsla['Daily_return'])
```

```python
# g. Calculate the volume-weighted average daily return.
tsla_a=tsla.dropna()
tsla_a.head()
```

```python
np.average(tsla_a['Daily_return'], weights=tsla_a['Volume'])
```

```python
np.mean((tsla_a['Daily_return']*tsla_a['Volume']).sum()/(tsla_a['Volume']).sum())
```

```python
np.sum(tsla_a['Daily_return']*tsla_a['Volume'])/(tsla_a['Volume']).sum()
```

```python
# h. Add a column containing the 1-week (5-day) moving average volume

# Rolling five-day mean
tsla['1-week (5-day) moving average volume']=tsla['Volume'].rolling(5).mean()

tsla
```

```python
# i. Add a column containing the 1-week (5-day) moving average daily return

tsla['1-week (5-day) moving average daily return']=tsla['Daily_return'].rolling(5).mean()

tsla.head(10)
```
