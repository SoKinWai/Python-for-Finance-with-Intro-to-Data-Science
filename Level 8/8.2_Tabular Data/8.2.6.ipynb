```python
import numpy as np
import pandas as pd
import yfinance as yf

data=pd.read_html('https://en.wikipedia.org/wiki/List_of_S%26P_500_companies')

tickers=sorted(data[0]['Symbol'])

#Show the first 10 sorted tickers
tickers[:10]
```

```python
# # Load two years of stock data for every ticker in the current S&P 500.
tickers = yf.Tickers(tickers)
SP500 = tickers.history(group_by='ticker',period='2y')
```

```python
SP500.head()
```

# a. ‘Stack’ the table to get it into row form, and demonstrate ‘unstack’ to get it back into the original form.


```python
SP500_a=SP500.stack(level=0).reset_index().rename(columns = {'level_1':'Ticker'})
SP500_a                           
```

```python
SP500_a.stack(level=0).unstack()
```

# b. Melt and unmelt the table (using pivot), as shown in the lecture.

```python
#Melt

melted=SP500_a.reset_index().melt(id_vars=['Date','Ticker'],var_name="Attribute",value_name="Value")
melted
```

```python
pivoted=melted.pivot(index=['Date','Ticker'],columns='Attribute')
pivoted = pivoted['Value'].reset_index()
pivoted
```

```python
pivoted.columns.name=None
pivoted
```

# d. Create a derived table (a.k.a a ‘summary table’), containing the total volume (in millions) per year, of each ticker.


```python
# synatx: data.groupby(data.date.dt.year)

SP500_a.groupby([SP500_a['Ticker'], SP500_a['Date'].dt.year]).agg({'Volume': 'sum'}).apply(lambda x:x*1e-6) 
```

# e. Create a derived table containing the total volume (in millions) of each ticker.

```python
SP500_a.groupby([SP500_a['Ticker']]).agg({'Volume': 'sum'}).apply(lambda x:x*1e-6) 
```

# f. Add a column to the original table containing the daily return per ticker (using the above formula).

```python
SP500_a['Daily return'] = (SP500_a['Close']-SP500_a['Close'].shift(1))/SP500_a['Close'].shift(1)
SP500_a
```

# g. Add a column to the original table containing the 1-week (5-day) moving average daily return, per ticker.

```python
SP500_a['1-week (5-day) moving average daily return'] = SP500_a['Daily return'].rolling(5).mean()
SP500_a
```

# h. Add a column to the original table containing the rolling cumulative sum of volume per ticker.


```python
SP500_a['Rolling cumulative sum of volume'] = SP500_a['Volume'].cumsum()
SP500_a
```
