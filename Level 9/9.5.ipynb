# This exercise uses the derived data created in 8.2.7. You should do the following

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


scaffold_Portfolio=pd.read_csv('scaffold_Portfolio.csv')
scaffold_Portfolio.reset_index()
del scaffold_Portfolio['Unnamed: 0']

scaffold_Portfolio["Date"] = pd.to_datetime(scaffold_Portfolio["Date"])
scaffold_Portfolio = scaffold_Portfolio.set_index(['Date'])


scaffold_Portfolio
```

```python
scaffold_Portfolio.dropna(subset=['Cumulative Daily P&L'], inplace=True)
scaffold_Portfolio
```

# a. Plot cumulative PnL for each ticker from your portfolio, on the same plot. Hint: Use DataFrame pivoting functionality.

```python
fig, ax1 = plt.subplots(figsize=(15,13))

for ticker in scaffold_Portfolio.Ticker.unique():
    data = scaffold_Portfolio[scaffold_Portfolio.Ticker == ticker]
    ax1.plot(data[['Cumulative Daily P&L']])

plt.grid()
plt.legend(list(scaffold_Portfolio.Ticker.unique()))

plt.title('Cumulative P&L for each ticker ', fontsize='20',fontweight ="bold")
plt.xlabel('Date', fontsize='15')
plt.ylabel('Cumulative P&L', fontsize='15')
plt.show()
```

```python
# Show the plot again via using DataFrame pivoting functionality 

p = pd.pivot_table(data=scaffold_Portfolio,
                                   index=['Ticker', 'Date'],
                                   values=['Cumulative Daily P&L'])

p = p.reset_index()
p = p.set_index('Date')
p
```

```python
fig, ax1 = plt.subplots(figsize=(15,13))

for ticker in p.Ticker.unique():
    data = p[p.Ticker == ticker]
    ax1.plot(data[['Cumulative Daily P&L']])

plt.grid()
plt.legend(list(p.Ticker.unique()))

plt.title('Cumulative P&L for each ticker ', fontsize='20',fontweight ="bold")
plt.xlabel('Date', fontsize='15')
plt.ylabel('Cumulative P&L', fontsize='15')
plt.show()
```

# b. Create a scatter plot of PnL per ticker, year 1 vs year 2. Use different color points for each ticker, and use alpha as appropriate

```python
scaffold_Portfolio.dropna(subset=['Daily P&L'], inplace=True)
scaffold_Portfolio
```

```python
scaffold_Portfolio=scaffold_Portfolio.reset_index()
scaffold_Portfolio.head()
```

```python
# I will choose AAPL, TSLA and CCL to show this plot.

fig, ax1 = plt.subplots(figsize=(10,5))


ticker_a = 'AAPL'
date_a = '2021-07-25'
y1 = scaffold_Portfolio[scaffold_Portfolio['Date']<=date_a].set_index('Date')
y2 = scaffold_Portfolio[scaffold_Portfolio['Date']>date_a].set_index('Date')
    
data_y1 = y1[y1.Ticker==ticker_a]
data_y2 = y2[y2.Ticker==ticker_a]
ax1.scatter(data_y1[['Daily P&L']], data_y2[['Daily P&L']],color='g', alpha=0.2)    
    

ticker_b = 'TSLA'
date_b = '2021-11-04'
y1 = scaffold_Portfolio[scaffold_Portfolio['Date']<=date_b].set_index('Date')
y2 = scaffold_Portfolio[scaffold_Portfolio['Date']>date_b].set_index('Date')
    
data_y1 = y1[y1.Ticker==ticker_b]
data_y2 = y2[y2.Ticker==ticker_b]
ax1.scatter(data_y1[['Daily P&L']], data_y2[['Daily P&L']],color='b', alpha=0.2)      
    
    
ticker_c = 'CCL'
date_c = '2021-12-06'
y1 = scaffold_Portfolio[scaffold_Portfolio['Date']<=date_c].set_index('Date')
y2 = scaffold_Portfolio[scaffold_Portfolio['Date']>date_c].set_index('Date')
    
data_y1 = y1[y1.Ticker==ticker_c]
data_y2 = y2[y2.Ticker==ticker_c]
ax1.scatter(data_y1[['Daily P&L']], data_y2[['Daily P&L']], color='orange',alpha=0.2)  
    


    
plt.legend([ticker_a,ticker_b,ticker_c])

plt.title('P&L per ticker, year 1 vs year 2 ', fontsize='20',fontweight ="bold")
plt.xlabel('Year 1 Daily P&L', fontsize='15')
plt.ylabel('Year 2 Daily P&L', fontsize='15')
plt.show()
```
