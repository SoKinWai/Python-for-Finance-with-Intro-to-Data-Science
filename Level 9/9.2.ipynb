# Create a DataFrame of 1,000,000 normally distributed random values. Plot a histogram using Pandas hist method. Test with several different values for bins â€“ how do different numbers of bins affect the shape of the chart?

```python
import numpy as np
import pandas as pd

# Create a DataFrame of 1,000,000 normally distributed random values.
s = np.random.normal(0, 0.1, 1000000)
len(s)
```

```python
df = pd.DataFrame({'Normally distributed random values':s})
df.head()
```

```python
#Plot a histogram using Pandas hist method.
df['Normally distributed random values'].hist(figsize=(10,5))
```

# Test with several different values for bins

```python
df['Normally distributed random values'].hist(figsize=(10,5), bins=50)
```

```python
df['Normally distributed random values'].hist(figsize=(10,5), bins=100)
```

```python
df['Normally distributed random values'].hist(figsize=(10,5), bins=200)
```

```python
df['Normally distributed random values'].hist(figsize=(10,5), bins=300)
```

```python
df['Normally distributed random values'].hist(figsize=(10,5), bins=500)
```

```python
df['Normally distributed random values'].hist(figsize=(10,5), bins=1000)
```
