```python
SP500=pd.read_csv('SP500.csv')
SP500.reset_index()

del SP500['Unnamed: 0']
del SP500['Adj Close']

SP500 = SP500.dropna()

SP500.head(10)
```

```python
from scipy.ndimage import gaussian_filter

# More fair smoothing
SP500['Smoothed 1-week (5-day) moving average daily return'] = gaussian_filter(SP500['1-week (5-day) moving average daily return'], 10)

SP500
```

```python
import plotly.graph_objs as go
from plotly.subplots import make_subplots

from ipywidgets import widgets

def tickerBrowser(df):
    
    fig = make_subplots(specs=[[{"secondary_y": True}]])

    search = widgets.Text(value = '', placeholder='Search string', description='', disabled=False, continuous_update=True)
    
    ticker = widgets.Dropdown(
        options=sorted(list(df['Ticker'].unique())),
        value=sorted(df['Ticker'].unique())[0],
        description='Ticker: ',
    )
    
    container = widgets.HBox(children=[search,ticker])

    # Assign an empty figure widget with two traces
    g = go.FigureWidget(fig, layout=go.Layout(
                            width=1000, height=500,
                            title=dict(
                                text='Daily return&Volume'
                            ),
                        ))
    
    
    g.add_scatter(name='Daily return')
    g.add_scatter(name='1-week (5-day) moving average daily return')
    g.add_scatter(name='Smoothed 1-week (5-day) moving average daily return')
    
    g.add_scatter(name='Rolling cumulative sum of volume', secondary_y=True)
    
    # Add figure title
    g.update_layout(
    title_text="Stock Daily Return & Cumulative Volume",
    )

    # Set x-axis title
    g.update_xaxes(title_text="Date")

    # Set y-axes titles
    g.update_yaxes(title_text="Daily <b>Return</b>", secondary_y=False)
    g.update_yaxes(title_text="Rolling Cumulative <b>Volume</b>", secondary_y=True)
    
    
    
    
    
        
    def response(change):
       
        temp_df = df[(df['Ticker'] == ticker.value) ]
        with g.batch_update():
            g.data[0].x = temp_df['Date'].tolist()
            g.data[0].y = temp_df['Daily return'].tolist()
            g.data[1].x = temp_df['Date'].tolist()
            g.data[1].y = temp_df['1-week (5-day) moving average daily return'].tolist()
            g.data[2].x = temp_df['Date'].tolist()
            g.data[2].y = temp_df['Smoothed 1-week (5-day) moving average daily return'].tolist()
            
            g.data[3].x = temp_df['Date'].tolist()
            g.data[3].y = temp_df['Rolling cumulative sum of volume'].tolist()
        
                
    
    def chTicker(change):
         if search.value:
            ticker.options = sorted(df[df['Ticker'].str.contains(search.value)]['Ticker'].unique())
         else:
            ticker.options = sorted(list(df['Ticker'].unique()))
            

    ticker.observe(response, names='value')
    search.observe(chTicker, names='value')


                       
    response(None)
    chTicker(None)
    
    display(container)
    display(g)
    
tickerBrowser(SP500)
```
