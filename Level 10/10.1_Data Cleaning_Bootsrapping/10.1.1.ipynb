# Perform data cleaning on the downloaded data. If the data is already clean, add a bunch of dummy ‘bad’ rows and columns which you can then demonstrate how to properly clean the data. Should perform at least the following (and other examples you can think of)

```python
import yfinance as yf
import pandas as pd

tsla = yf.Ticker("TSLA")
tslaData = tsla.history(period="max")
tslaData['Ticker'] = 'TSLA'

aapl = yf.Ticker("AAPL")
aaplData = aapl.history(period="max")
aaplData['Ticker'] = 'AAPL'

allData = pd.concat([tslaData, aaplData], axis=0)
allData.iloc[-2,-1] = 'aapl'
allData.iloc[-3,-1] = ' AAPL '
allData = allData.append(allData.iloc[-1,:])
allData = allData.append(allData.iloc[-5,:])
allData.iloc[-1,-1] = 'MSFT'
```

```python
allData.iloc[-100, -4] = None
allData.iloc[50, -4] = None
```

# a. Drop Nulls


```python
allData.dropna(inplace=True)
allData
```

# b. Remove irrelevant columns

```python
allData = allData.drop(['Open','High','Low'], axis=1) # inplace
allData
```

# c. Standardize a date/time column

```python
allData = allData.reset_index()
allData['Date_time'] = pd.to_datetime(allData['Date'])  
allData['Date_time'] = allData['Date_time'].dt.strftime('%Y-%m-%d')  
allData
```

# f. Winsorize outliers using both clip and a winsorize function.

I will show few ways to remove outliers.

```python
allData["Volume"].describe()
```

```python
print(f'Count: {allData["Volume"].count():,}')     # Number of rows
print(f'Average: {allData["Volume"].mean():,}')    
print(f'Min: {allData["Volume"].min():,}')
print(f'Max: {allData["Volume"].max():,}')
print(f'5%: {allData["Volume"].quantile(0.05):,}')
print(f'95%: {allData["Volume"].quantile(0.95):,}')

#These are called outliers. 
```

```python
# Remove volume outliers

volMin = allData["Volume"].quantile(0.05)
volMax = allData["Volume"].quantile(0.95)
allDataNoOutliers = allData[allData.Volume.between(volMin, volMax)]
```

```python
print(f'Count: {allDataNoOutliers["Volume"].count():,}')
print(f'Average: {allDataNoOutliers["Volume"].mean():,}')
print(f'Min: {allDataNoOutliers["Volume"].min():,}')
print(f'Max: {allDataNoOutliers["Volume"].max():,}')
print(f'5%: {allDataNoOutliers["Volume"].quantile(0.05):,}')
print(f'95%: {allDataNoOutliers["Volume"].quantile(0.95):,}')
```

```python
# Clip

volMin = allData["Volume"].quantile(0.05)
volMax = allData["Volume"].quantile(0.95)

allDataClipped = allData.copy()
allDataClipped['Volume'] = allDataClipped['Volume'].clip(volMin, volMax)
```

```python
print(f'Count: {allDataClipped["Volume"].count():,}')
print(f'Average: {allDataClipped["Volume"].mean():,}')
print(f'Min: {allDataClipped["Volume"].min():,}')
print(f'Max: {allDataClipped["Volume"].max():,}')
print(f'5%: {allDataClipped["Volume"].quantile(0.05):,}')
print(f'95%: {allDataClipped["Volume"].quantile(0.95):,}')


# The result it much better!!!
```

```python
# Winsorize

import scipy.stats.mstats

allDataWinsorized = allData.copy()
allDataWinsorized["Volume"] = scipy.stats.mstats.winsorize(allDataWinsorized["Volume"], [.05, .05])
```

```python
print(f'Count: {allDataWinsorized["Volume"].count():,}')
print(f'Average: {allDataWinsorized["Volume"].mean():,}')
print(f'Min: {allDataWinsorized["Volume"].min():,}')
print(f'Max: {allDataWinsorized["Volume"].max():,}')
print(f'5%: {allDataWinsorized["Volume"].quantile(0.0):,}')
print(f'95%: {allDataWinsorized["Volume"].quantile(0.95):,}')
```
